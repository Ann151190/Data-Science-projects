```
def features(sent,i):
  features = {
              'bias': 1.0,
              'token.lower()': sent.lower_,
              'token.suffix': sent.suffix_,
              'token.prefix': sent.prefix_,
              'token.isupper()': sent.is_upper,
              'token.istitle()': sent.is_title,
              'token.isdigit()': sent.is_digit,
              }
  if i == 0:
    features['BOS'] = True
  if i > len(sent):
    features['EOS'] = True
  if i < len(sent): 
    features.update({
                    '+1:token.lower()': sent.nbor().lower_,
                    '+1:token.istitle()': sent.nbor().is_title,
                    '+1:token.isupper()': sent.nbor().is_upper,
                    }) 
  if i != 0: 
    features.update({
                    '-1:token.lower()': sent.nbor(-1).lower_,
                    '-1:token.istitle()': sent.nbor(-1).is_title,
                    '-1:token.isupper()': sent.nbor(-1).is_upper,
                    })
  return features
```
