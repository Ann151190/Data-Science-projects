```
country_unique =list(df2['Country/Region'].unique())
```

```
for i in country_unique:
  with open('country.pickle'+ str(i), 'wb') as f:
    pickle.dump(df2.loc[(df2['Country/Region'] == i)], f)
```

```
with open('country.pickleMainland China', 'rb') as f:
    df2_china = pickle.load(f)
    print(df2_china.shape)
```

```
df2_china.dtypes
```

```
column_to_float = ['Confirmed', 'Deaths', 'Recovered']
column_to_date = ['ObservationDate', 'Last Update']

for col in column_to_float:
  df2_china[col] = df2_china[col].astype(np.float32)

for col in column_to_date:
  df2_china[col] = pd.to_datetime(df2_china[col], errors='coerce')
```
